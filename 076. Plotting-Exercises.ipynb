```python
import numpy as np
import math

x = np.linspace(0,4 * np.pi, 100)
y = np.sin(x) + np.sin(2*x)

xr = np.around(x, decimals=2)
yr = np.around(y, decimals=2)

print('X Values')
print(xr)
print('Y Values')
print(yr)
```

```python
plt.plot(xr, yr)
plt.scatter(xr, yr, color='red')
```
