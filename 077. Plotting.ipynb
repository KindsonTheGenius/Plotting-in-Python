```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2 * np.pi, 40)
x = np.sin(t)
y = np.sin(t) * np.cos(t)
```

```python
# plt.subplot(2,1,1)
plt.plot(x, y, 
         markersize = 10,
         linewidth = 15,
         color = 'green',
        )

plt.show()
```
